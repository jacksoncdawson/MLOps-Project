## Lookup Table Transformation

I think a lot of these columns have a string value, and a numeric value. The numeric value is what we will want to use for our model, so we will need to create a lookup table for the string values.

Below I'll check for unique values in each column, to see if likely pair columns are consistent.

```python
pairs = [
    ('product_code', 'prod_name'),
    ('product_type_no', 'product_type_name'),
    ('graphical_appearance_no', 'graphical_appearance_name'),
    ('colour_group_code', 'colour_group_name'),
    ('perceived_colour_value_id', 'perceived_colour_value_name'),
    ('perceived_colour_master_id', 'perceived_colour_master_name'),
    ('department_no', 'department_name'),
    ('index_code', 'index_name'),
    ('index_group_no', 'index_group_name'),
    ('section_no', 'section_name'),
    ('garment_group_no', 'garment_group_name'),
]

for pair in pairs:
    print(pair[0], pair[1])

    left = np.sort(df_transactions[pair[0]].value_counts().to_numpy())
    right = np.sort(df_transactions[pair[1]].value_counts().to_numpy())
    
    if np.array_equal(left, right):
        print("PAIR")
    else:
        print("NOT PAIR")
```

```python
for a, b in pairs:
    vc_a = df_transactions[a].value_counts()
    vc_b = df_transactions[b].value_counts()

    if not np.array_equal(np.sort(vc_a.to_numpy()), np.sort(vc_b.to_numpy())):
        print(f"\nOFF-PAIR: {a} vs {b}")
        print("unique values:", vc_a.shape[0], vc_b.shape[0])
        print("top counts A:", vc_a.head().to_dict())
        print("top counts B:", vc_b.head().to_dict())

        print("count-of-counts A:")
        print(vc_a.value_counts().sort_index().head(10))

        print("count-of-counts B:")
        print(vc_b.value_counts().sort_index().head(10))
```

```python
off_pairs = [
    ('product_code', 'prod_name'),
    ('product_type_no', 'product_type_name'),
    ('department_no', 'department_name'),
    ('section_no', 'section_name'),
]


def ambiguous_examples(df, left_col, right_col, limit=5):
    left_to_right = (
        df[[left_col, right_col]]
        .drop_duplicates()
        .groupby(left_col)[right_col]
        .agg(lambda s: sorted(s.dropna().astype(str).unique().tolist()))
    )
    right_to_left = (
        df[[left_col, right_col]]
        .drop_duplicates()
        .groupby(right_col)[left_col]
        .agg(lambda s: sorted(s.dropna().astype(str).unique().tolist()))
    )

    left_ambiguous = left_to_right[left_to_right.apply(len) > 1].head(limit)
    right_ambiguous = right_to_left[right_to_left.apply(len) > 1].head(limit)

    return left_ambiguous, right_ambiguous


for left_col, right_col in off_pairs:
    print(f"\n{left_col} vs {right_col}")

    left_ambiguous, right_ambiguous = ambiguous_examples(
        df_transactions, left_col, right_col
    )

    if not left_ambiguous.empty:
        print(f"{left_col} values with multiple {right_col} values:")
        display(left_ambiguous.to_frame(name=f"{right_col}_examples"))

    if not right_ambiguous.empty:
        print(f"{right_col} values with multiple {left_col} values:")
        display(right_ambiguous.to_frame(name=f"{left_col}_examples"))
```

| Column A        | Column B          | Pair Status | Notes Notes                                                                                                                                                                              | Solution |
| --------------- | ----------------- | ----------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | -------- |
| product_code    | prod_name         | Not Pair    | `product_code` can have multiple `prod_name` spellings, and some `prod_name` values are reused across multiple codes. Keep `product_code` as the lookup key and choose a canonical name. |
| product_type_no | product_type_name | Not Pair    | `Umbrella` appears under two different `product_type_no` values (`83` and `532`), so the name is not unique.                                                                             |
| department_no   | department_name   | Not Pair    | Several `department_name` values map to multiple `department_no` values, including `Jersey`, `Belts`, and `Swimwear`. Use `department_no` as the stable key.                             |
| section_no      | section_name      | Not Pair    | `Ladies Other` maps to two `section_no` values (`4` and `17`), so `section_name` alone is not unique.                                                                                    |


```python
def normalize_to_most_prevalent(df, group_col, value_col):
    pair_counts = (
        df.groupby([group_col, value_col], dropna=False)
        .size()
        .reset_index(name="row_count")
    )
    pair_counts["_tie_break"] = pair_counts[value_col].astype(str)

    canonical = (
        pair_counts.sort_values(
            [group_col, "row_count", "_tie_break"],
            ascending=[True, False, True],
            kind="mergesort",
        )
        .drop_duplicates(subset=[group_col], keep="first")
        [[group_col, value_col, "row_count"]]
        .rename(
            columns={
                value_col: f"canonical_{value_col}",
                "row_count": "canonical_row_count",
            }
        )
    )

    option_counts = (
        pair_counts.groupby(group_col, dropna=False)[value_col]
        .nunique(dropna=False)
        .rename("n_options")
        .reset_index()
    )
    ambiguous_groups = option_counts[option_counts["n_options"] > 1][
        [group_col, "n_options"]
    ]

    updates = ambiguous_groups.merge(canonical, on=group_col, how="left")
    canonical_col = f"canonical_{value_col}"
    update_map = updates.set_index(group_col)[canonical_col]

    original = df[value_col].copy()
    mask = df[group_col].isin(update_map.index)
    df.loc[mask, value_col] = df.loc[mask, group_col].map(update_map)

    changed_rows = (
        ~original.eq(df[value_col]) & ~(original.isna() & df[value_col].isna())
    ).sum()

    return (
        updates.sort_values(["n_options", "canonical_row_count"], ascending=[False, False]),
        int(changed_rows),
    )


# Work on a cleaned copy so the original frame is still available for comparison.
df_transactions_clean = df_transactions.copy()
normalization_log = []

for left_col, right_col in off_pairs:
    for group_col, value_col in [(left_col, right_col), (right_col, left_col)]:
        updates, changed_rows = normalize_to_most_prevalent(
            df_transactions_clean, group_col, value_col
        )
        normalization_log.append(
            {
                "group_col": group_col,
                "value_col": value_col,
                "groups_normalized": len(updates),
                "rows_changed": changed_rows,
            }
        )

        print(f"\n{group_col} -> {value_col}")
        print(f"groups normalized: {len(updates)}")
        print(f"rows changed: {changed_rows}")
        if not updates.empty:
            display(updates.head(10))

normalization_summary = pd.DataFrame(normalization_log)
normalization_summary
```
